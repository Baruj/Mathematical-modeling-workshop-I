## Tarea 1

Clasificación de calificaciones

Escribe un programa en Python que solicite al usuario una calificación numérica entre 0 y 10 y la clasifique de la siguiente forma:


| Rango                  | Clasificación |
| ---------------------- | ------------- |
| Menor que 6            | `NA`          |
| De 6 a menor que 7.6   | `S`           |
| De 7.6 a menor que 8.6 | `B`           |
| De 8.6 a 10            | `MB`          |

El programa debe:

Pedir la calificación usando input().
Convertir el valor ingresado a tipo float.
Usar if, elif y else.
Imprimir la clasificación correspondiente.
Si la calificación es menor que 0 o mayor que 10, imprimir: "Calificación no válida"

```python
# Programa que solicida una calificación entre 0 y 10 y la clasifica
# Autor: Baruj Furlong Oceguera
# Fecha 11 de mayo del 2026

# Se pide la calificación al usuario
grade = float(input("Ingresa la calificación: "))
if grade < 0 or grade > 10 :
    print("Calificación no válida")
elif grade >=0 and grade < 6:
    print('NA')
elif grade >=6 and grade < 7.6 :
    print('S')
elif grade >= 7.6 and grade < 8.6:
    print('B')
elif grade >= 8.6 and grade <= 10:
    print('MB')
    
```

## Tarea 2

Análisis de una lista de números

Considera la siguiente lista:
numeros = [4, 7, 12, 15, 18, 21, 24, 30, 33, 40]

Escribe un programa que haga lo siguiente:

Imprima todos los elementos de la lista usando un ciclo for.
Calcule la suma de todos los elementos.
Calcule cuántos números son pares.
Calcule cuántos números son impares.
Construya una nueva lista llamada mayores_20 que contenga solamente los números mayores que 20.
Imprima los resultados.

```python
numeros = [4, 7, 12, 15, 18, 21, 24, 30, 33, 40]
suma=0
pares=0
impares=0
mayores_20 = []
print("Elementos de la lista original")
for i in numeros:
    print(i)
    if i%2==0: pares+=1
    else: impares+=1
    if i>20:
        mayores_20.append(i)
    suma+=i

print("La suma es "+str(suma))
print(str(pares)+" números son pares")    
print(str(impares)+" números son impares")    
print("A continuacion se muestra la lista con valores mayores a 20: "+str(mayores_20))
```
